```python
# Cell 1: Import libraries and connect to the database
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a nice style for the charts
sns.set_theme(style="whitegrid")

# Connect to your database file
conn = sqlite3.connect('MovieLibrary.db')

print("✅ Libraries imported and database connected!")
```

```python
# Cell 2: Query for the top 10 directors and load the data
query = """
    SELECT
        d.director_name,
        COUNT(m.movie_id) AS movie_count
    FROM Movies m
    JOIN Directors d ON m.director_id = d.director_id
    GROUP BY d.director_name
    ORDER BY movie_count DESC
    LIMIT 10;
"""
top_directors_df = pd.read_sql_query(query, conn)

# Display the table of top directors
top_directors_df
```

```python
# Cell 3: Create a bar chart to visualize the results
plt.figure(figsize=(12, 8))
sns.barplot(x='movie_count', y='director_name', data=top_directors_df, palette='viridis')
plt.title('Top 10 Most Prolific Directors in IMDb Top 1000', fontsize=16)
plt.xlabel('Number of Movies', fontsize=12)
plt.ylabel('Director', fontsize=12)
plt.show()
```
